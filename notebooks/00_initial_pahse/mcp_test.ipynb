```python
from fastmcp.client import Client
from rich.pretty import pprint

client = Client(transport='http://localhost:8000/sse')
async with client:
	tools = await client.list_tools()
	pprint(tools)
```

```python
async with client:
    res = await client.call_tool('addition_tool', {'a': 5, 'b': 10})
    pprint(res)
```

```python
from agent_framework.tools.mcp_tool import MCPTool, MCPClient
from rich.pretty import pprint

client = MCPClient()
await client.connect_sse('http://localhost:8000/sse')
async with client:
    res = await client.call_tool('addition_tool', {'a': 5, 'b': 10})
    pprint(res)

    res_search = await client.call_tool('web_search_tool', {'url': 'https://ai-sdk.dev/docs/ai-sdk-ui/chatbot',
                                                            'user_agent': 'windows',
                                                            })
    pprint(res_search)
```
